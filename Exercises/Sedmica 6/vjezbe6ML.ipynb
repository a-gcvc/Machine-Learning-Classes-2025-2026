```python
freq_table = pd.crosstab(df['coffee_name'], 'no_of_coffee_types', normalize=True) #RELATIVE FREQUENCY TABLE: postotak svih pojedinačnih jedinstvenih vrijednosti kategoričke varijable u datasetu
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Time_of_Day'], normalize="index") #two-way frequency table: prikaz dvije varijable po redovima i kolonama
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Weekday'])
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Weekday'], normalize="index")
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Weekday'], normalize="columns")
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Month_name'])
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Month_name'], normalize=True)
freq_table
```

```python
freq_table = pd.crosstab(df['coffee_name'], df['Month_name'], normalize="columns")
freq_table
```

```python
plt.hist(df['Weekdaysort'])
plt.title("Profit per week day")
plt.xlabel("Day in week")
plt.ylabel("Profit")
plt.show()
```

```python
#kvartal
q1=df['money'].quantile(0.25) #prvi kvartal
q3=df['money'].quantile(0.75) #treći kvartal
iqr=q3-q1

print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")
```

```python
plt.hist(df['money'])
plt.axvline(q1, color='r', linestyle='--')
plt.axvline(q3, color='r', linestyle='--')
plt.title("Profit")
plt.show()
```

```python
#mjere distribucije
print(df['money'].mean())
print(df['money'].median())
print(df['money'].mode())
print(df['money'].var()) #varijansa
print(df['money'].std()) #standardna devijacija
print(df['money'].skew()) #skewness: asimetrija
print(df['money'].kurt()) #kurtosis: zakrivljenost
```
