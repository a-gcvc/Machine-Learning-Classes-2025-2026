Kolone Student_ID, Affects_Academic_Performance i Overall_Imapct su izbačene iz dataset-a.
Kolona **Student_ID** nije relevantna za bilo kakav oblik razmatranja prilikom analize, a kasnije i treniranja podataka, dok kolone **Affects_Academic_Performance** i **Overall_Imapct** se odnose na one kolone na osnovu kojih bi trebali dobiti rezultate treniranja.

```python
df.drop(columns=['Student_ID', 'Affects_Academic_Performance', 'Overall_Impact'], inplace=True)
```

Prikazani su svi nazivi kolona koje nisu numeričke:

```python
categorical_col = df.select_dtypes("object").columns
categorical_col
```

Na osnovu kolona koje nisu numeričke je kreiran zaseban dataframe:

```python
df_categoric = df[categorical_col]
```

Prikaz prvih 10 kolona novog dataframe-a:

```python
df_categoric.head(10)
```

U nastavku su kolone koje nisu numeričke konvertovane u numeričke varijable:

```python
encoder = LabelEncoder()
```

```python
df['Gender'].unique()
```

```python
df_categoric['Gender'] = encoder.fit_transform(df_categoric['Gender'])
df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
df['Academic_Level'].unique()
```

```python
df_categoric['Academic_Level']= df_categoric['Academic_Level'].map({'High School':0, 'Undergraduate':1, 'Graduate':2})
df['Academic_Level'] = df['Academic_Level'].map({'High School':0, 'Undergraduate':1, 'Graduate':2})
```

```python
df['Country'].unique()
```

```python
df_categoric['Country'] = encoder.fit_transform(df_categoric['Country'])
df['Country'] = encoder.fit_transform(df['Country'])
```

```python
df.head()
```

```python
df['Most_Used_Platform'].unique()
```

```python
df_categoric['Most_Used_Platform'] = encoder.fit_transform(df_categoric['Most_Used_Platform'])
df['Most_Used_Platform'] = encoder.fit_transform(df['Most_Used_Platform'])
```
