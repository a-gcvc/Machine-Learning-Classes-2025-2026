```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('student-mat.csv', sep=';')

print(df.columns)
```

```python

sns.set_theme(style="darkgrid")

plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=8, kde=True, color='teal')

plt.title('Distribucija godina učenika', fontsize=15)
plt.xlabel('Godine', fontsize=12)
plt.ylabel('Broj učenika', fontsize=12)

plt.show()
```
