```python
# zadatak 5 - 09.04.2026

#analizirati da li su varijable iskrivljene ili nisu
#za numeričke varijable izračunati:
# standardna devijacija
# varijansa
# skew
# kurtosis
# aritmetička sredina
# koeficijent varijacije: pomjer standardne devijacije i aritmetičke sredine pomnožen sa 100
# kvartili q1 i q3
# prikazati prethodne varijable grafički
# po obliku distribucije vidjeti da li su normalno distribuirane ili ne

# nacrtati histograme za numeričke varijable
```

```python
print(df['Age'].std()) #standardna devijacija
print(df['Age'].var()) #varijansa
print(df['Age'].skew()) #skewness: asimetrija
print(df['Age'].kurt()) #kurtosis: zakrivljenost
print(df['Age'].mean()) #aritmetička sredina
koef_var = ((df['Age'].std()) / (df['Age'].mean())) * 100
print (koef_var)

q1 = df['Age'].quantile(0.25) #prvi kvartil
q3 = df['Age'].quantile(0.75) #treći kvartil

print("Q1:", q1)
print("Q3:", q3)

print(trim_mean(df['Age'].dropna(), 0.05)) # skraćena aritmetička sredina
mean_age = df['Age'].mean()
trimmed_age = trim_mean(df['Age'].dropna(), 0.05)
print("Age - Mean:", mean_age)
print("Age - Trimmed Mean:", trimmed_age)
print("Age - Razlika:", mean_age - trimmed_age)
```

```python
df['Age'].hist(bins=10)

plt.xlabel("Age")
plt.ylabel("Frekvencija")
plt.title("Histogram varijable Age")

plt.show()
```
